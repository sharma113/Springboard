```python
# Let's import the libraries. 
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from scipy.stats import chi2_contingency
```

```python
datafile_atypical='atypical.csv'
df_atypical = pd.read_csv(datafile_atypical)
```

```python
datafile_typical='typical.csv'
df_typical = pd.read_csv(datafile_typical)
```

The target variable will be whether the patient suffered from a myocardial complication.  We will comparing the different features based on the two different populations of complication and non-complication. 

```python
#First we will look at the 'Age' feature in the two data sets. 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))

df_typical['AGE'].sort_values().hist(ax=ax1)
ax1.set_title('Age in Non-Complicated Patients')
ax1.set_xlabel('Age')
ax1.set_ylabel('Patient Count')

df_atypical['AGE'].sort_values().hist(ax=ax2)
ax2.set_title('Age in Complicated Patients')
ax2.set_xlabel('Age')
ax2.set_ylabel('Patient Count')
plt.show()
```

```python
#Perform t-test and p-value on above. 
t_value, p_value = stats.ttest_ind(df_typical['AGE'], df_atypical['AGE'])
print("AGE: t-value = {:.2f}, p-value = {:.2f}".format(t_value, p_value))
print('Mean Non-Complicated:',df_typical['AGE'].mean(),'Mean Complicated:',df_atypical['AGE'].mean())
```

There appears to be a significant age difference between patients with a heart complication and those without.

```python
#We will look at the 'SEX' feature in the two data sets. 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))

df_typical['SEX'].sort_values().hist(ax=ax1)
ax1.set_title('Gender in Non-Complicated Patients')
ax1.set_xlabel('Gender')
ax1.set_ylabel('Patient Count')

df_atypical['SEX'].sort_values().hist(ax=ax2)
ax2.set_title('Gender in Complicated Patients')
ax2.set_xlabel('Gender')
ax2.set_ylabel('Patient Count')
plt.show()
```

```python
contingency_table = pd.crosstab(df_typical['SEX'], df_atypical['SEX'])
```

```python
# Perform the chi-squared test and print the results
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-squared statistic = {:.2f}".format(chi2))
print("p-value = {:.4f}".format(p))
```

In this case 0 = Female and 1= Male.  Give a p-value of .3841 there is no signifcant gender differnce between the two populations.

```python
#Perform Chi Squared to get P-values on all features.

stat_dict={}
for i in column_names:
    # Make contingency table
    contingency_table = pd.crosstab(df_typical[i], df_atypical[i])
    # Perform the chi-squared test and print the results
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    key = i
    value_1 = p
    value_2 = chi2
    stat_dict.update({key: (value_1, value_2)})
```

```python
#Loop throught the features that significant different
for key,values in stat_dict.items():
    value_1, value_2 = values
    if value_1 <= 0.05:
        print(key,value_1)
```
