```python
#Convert Systolic and Diastolic ICU readings into bins based on American Heart Association guidelines for labeling blood pressure.

# Define the bin edges and labels for systolic HTN
systolic_bins = [0, 120, 130, 140, 180, np.inf]
systolic_labels = [0, 1, 2, 3, 4]

# Define the bin edges and labels for diastolic HTN
diastolic_bins = [0, 80, 90, 120, np.inf]
diastolic_labels = [0, 1, 2, 3]

# Binning for systolic HTN
df['S_AD_ORIT'] = pd.cut(df['S_AD_ORIT'], bins=systolic_bins, labels=systolic_labels)

# Binning for diastolic HTN
df['D_AD_ORIT'] = pd.cut(df['D_AD_ORIT'], bins=diastolic_bins, labels=diastolic_labels)
```

```python
#Here we are filling in the null values based on the mean value of that column based on the gender and age.
for i in column_names:
    # Filter rows with null values in column
    null_rows = df[df[i].isnull()]

        # Loop through null rows
    for index, row in null_rows.iterrows():
        age = row['AGE']
        sex = row['SEX']
    
        # Calculate mean of column for corresponding values in 'AGE' and 'SEX'
        mean_value = df[(df['AGE'] == age) & (df['SEX'] == sex)][i].notnull().mean()
        # Print the mean value
        #print(f"Column={i} and Mean value for AGE={age} and SEX={sex}: {mean_value}")
        
        #Update Null Values with rounding of the mean value based on Age and Sex.
        df.loc[index, i] = round(mean_value)
```
